```python
import pandas as pd
import numpy as np
```

```python
SleepDay = pd.read_csv("/content/drive/MyDrive/week_8/sleepDay_merged.csv")
minuteSleep = pd.read_csv("/content/drive/MyDrive/week_8/minuteSleep_merged.csv")
```

```python
Sleep_merge_df = pd.merge(SleepDay,minuteSleep,on='Id')
```

```python
Sleep_merge_df.head(5)
```

Cleaning the data

```python
Sleep_merge_df.info()
```

```python
Sleep_merge_df['SleepDay'] = pd.to_datetime(Sleep_merge_df['SleepDay']).dt.strftime("%m/%d/%Y")
Sleep_merge_df["SleepDay"] = Sleep_merge_df["SleepDay"].str.replace("-","/")
```

```python
#convert into CSV file
Sleep_merge_df.to_csv("/content/drive/MyDrive/omer_8/Sleep_Data.csv")
```
