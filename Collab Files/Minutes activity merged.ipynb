```python
import pandas as pd
import numpy as np
```

```python
Intensities = pd.read_csv("/content/drive/MyDrive/week_8/minuteIntensitiesNarrow_merged.csv")
Calories = pd.read_csv("/content/drive/MyDrive/week_8/minuteCaloriesNarrow_merged.csv")
METs = pd.read_csv("/content/drive/MyDrive/week_8/minuteMETsNarrow_merged.csv")
Steps = pd.read_csv("/content/drive/MyDrive/week_8/minuteStepsNarrow_merged.csv")
```

```python
merge_df = pd.merge(Intensities,Calories,on=['Id','ActivityMinute'])
merge_df2 = pd.merge(merge_df,METs,on=['Id','ActivityMinute'])
Minute_merge_df = pd.merge(merge_df2,Steps,on=['Id','ActivityMinute'])
```

```python
Minute_merge_df.head(5)
```

Data Cleaning

```python
Minute_merge_df['ActivityMinute'] = pd.to_datetime(Minute_merge_df['ActivityMinute']).dt.strftime("%m/%d/%Y")
Minute_merge_df['ActivityMinute'].str.replace("-","/")
```
