Data Cleaning


```python
del data["LoggedActivitiesDistance"]
```

```python
del data["SedentaryActiveDistance"]
```

```python
data.head(5)
```

```python
#conver ActivityDate and time format
data["ActivityDate"] = pd.to_datetime(data["ActivityDate"])
data["ActivityDate"] = pd.to_datetime(data["ActivityDate"]).dt.strftime("%m/%d/%Y")
```

```python
data["ActivityDate"] = data["ActivityDate"].str.replace("-","/")
data["ActivityDate"]
```

Exploratory Data Analysis (EDA)

```python
#histogram for Steps
plt.figure(figsize=(10,6))
#using seaborn to create a histogram
sns.histplot(data['TotalSteps'])
#add title and labels to the plot
plt.title('Distribution of Total Steps')  # Title of the plot
plt.xlabel('Steps')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis

# Display the plot
plt.show()
```
