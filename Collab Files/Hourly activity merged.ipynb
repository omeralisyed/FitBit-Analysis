```python
import pandas as pd
import numpy as np
```

```python
calories = pd.read_csv("/content/drive/MyDrive/week_8/hourlyCalories_merged.csv")
Intesities = pd.read_csv("/content/drive/MyDrive/week_8/hourlyIntensities_merged.csv")
steps = pd.read_csv("/content/drive/MyDrive/week_8/hourlySteps_merged.csv")
```

```python
merge_df = pd.merge(calories,Intesities, on=['Id','ActivityHour'])
Hourly_merge_df = pd.merge(merge_df,steps,on=['Id','ActivityHour'])

#display the merged dataframe
Hourly_merge_df.head()
```

Cleaning Data

```python
Hourly_merge_df.isnull().sum()
```

```python
Hourly_merge_df.describe()
```

```python
#cleaning the Activity Hour data
Hourly_merge_df["ActivityHour"] = pd.to_datetime(Hourly_merge_df["ActivityHour"]).dt.strftime('%m/%d/%Y')
Hourly_merge_df["ActivityHour"].str.replace("-","/")
```

Save new DataFrame to a new CSVfile


```python
Hourly_merge_df.to_csv("/content/drive/MyDrive/Hourly_Data.csv")
```
